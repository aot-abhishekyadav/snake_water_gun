```python
# Rock Paper scissor game
import random
def play():
    choices = ["rock","paper","scissor"]
    computer = random.choice(choices)
    print("Your choices are:\nRock\nPaper\nScissor")
    your = input("Enter your choice : ").lower()

    print(f"your choice is : {your}")
    print(f"computer choice is : {computer}")


    if(your == computer):
        print("Draw")

    else :
        if(your == "rock" and computer == "scissor"):
            print("You win")

        elif(your == "scissor" and computer == "paper"):
            print("You win")

        elif(your == "paper" and computer == "rock"):
            print("You win")

        elif (your == "rock" and computer == "paper"):
            print("You lose")

        elif(your == "scissor" and computer == "rock"):
            print("You lose")

        elif(your == "paper" and computer == "scissor"):
            print("You lose")

        else :
            print("Invalid choice choose between Rock, Paper, Scissor")
play()
```
